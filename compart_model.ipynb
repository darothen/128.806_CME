```python
y0 = [612, 730, 140, 37000, 580, 1500, 5300, 1.0]
emis_func = lambda t: emissions_22p6(t, 100.)
model = CarbonModel(*y0, emis_func=emis_func)

output = model.integrate(odeint, 150.)
```

```python
#output[['M1', 'M2', 'M3', 'M4', 'M5', 'M6']].plot()
#plt.semilogy()
#plt.twinx()
output['atm_ppm'].plot(style='k', label="atm CO2")
plt.xlim(1850, 1959)
```

```python
print "Euler"
o_euler = model.integrate_odespy(Euler, 150., dt=1.)
print "Backwards Euler"
o_euler_bk = model.integrate_odespy(BackwardEuler, 150., dt=1./10.)[::10]
print "RK4"
o_RK4 = model.integrate_odespy(RK4, 150., dt=1.)
print "Vode"
o_best = model.integrate_odespy(Vode, 150., dt=1.)

o_euler.atm_ppm.plot(label="Euler")
o_euler_bk.atm_ppm.plot(marker='o', ms=5, label="BackwardEuler")
o_RK4.atm_ppm.plot(label="RK4")
o_best.atm_ppm.plot(style="--k", zorder=-999, lw=8, label="Vode")
plt.legend()
```

```python
t0 = 1850
t_break = 2050
t = np.arange(0., t_break - t0, 1.)
emis_early = np.array([emissions_22p6(ti) for ti in t])

E_2050 = np.array(emissions_22p6(2050-1850.))
f = lambda t, lam, t_off=t_break-t0: E_2050*np.exp(-(t-t_off)/lam)
t_final = 2100 - t0
t_late = np.arange(t_break - t0, t_final, 1.)
emis_late = np.array([f(ti, 20.) for ti in t_late])

plt.plot(t, emis_early)#, color='rgk')
plt.plot(t_late, emis_late)#, color='rgk')

print emis_early[-1]
print emis_late[-1]
```

```python
def cust_emis(t):
    lam = 20.
    t_off = 2050-1850
    if t <= t_off:
        return emissions_22p6(t)
    else:
        E_2050 = np.array(emissions_22p6(t_off))
        return E_2050*np.exp(-(t-t_off)/lam)
    
ts = np.arange(0., 2100-1850., 1.)
ee = np.array([cust_emis(ti) for ti in ts])
plt.plot(ts, ee)

plt.figure()
model = CarbonModel(*y0, emis_func=cust_emis)

output = model.integrate(odeint, 1000.)
output.atm_ppm.plot()
```
